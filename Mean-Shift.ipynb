```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

#3D plot
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
```

```python
centers = [[2,2,2],[7,7,7],[5,13,13]]
#create data
```

```python
#Skip the Y by _ since we don't use or have target(y) in clustering
X, _ = make_blobs(n_samples=150,centers=centers,cluster_std=0.5)
```

```python
X
```

```python
#Plot 3D
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2])
plt.show()
```

```python
#Mean shift clustering
from sklearn.cluster import MeanShift
ms = MeanShift()
ms.fit(X)
```

```python
y_pred = ms.fit_predict(X)
y_pred
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c=y_pred,cmap='rainbow')
plt.show()
```
