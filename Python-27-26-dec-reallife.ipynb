```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sns.set_style('whitegrid')
```

```python
data = pd.read_csv('reallifedata.csv')
```

```python
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv
```

```python
data_no_mv.describe(include='all')
```

```python
data_no_mv.isnull().sum()
```

```python
data_no_mv = data_no_mv.drop(['Model'],axis=1)
```

```python
data_no_mv
```

```python
sns.distplot(data_no_mv['Price'])
plt.show()
```

```python
q = data_no_mv['Price'].quantile(0.99)
```

```python
q
```

```python
data_1 = data_no_mv[data_no_mv['Price']<q]
```

```python
data_1.describe(include='all')
```

```python
sns.distplot(data_1['Price'])
plt.show()
```

```python
sns.distplot(data_no_mv['Mileage'])
plt.show()
```

```python

q = data_1['Mileage'].quantile(0.99)
```

```python
q
```

```python
data_2 = data_1[data_1['Mileage']<q]
```

```python
data_2.describe(include='all')
```

```python
sns.distplot(data_2['Mileage'])
plt.show()
```

```python
sns.distplot(data_no_mv['EngineV'])
plt.show()
```

```python
data_3 = data_2[data_2['EngineV']<6.5]
```

```python
data_3.describe(include='all')
```

```python
sns.distplot(data_3['EngineV'])
plt.show()
```

```python
sns.distplot(data_no_mv['Year'])
plt.show()
```

```python
q = data_3['Year'].quantile(0.01)
```

```python
q
```

```python
data_4 = data_3[data_3['Year']>q]
```

```python
data_4.describe(include='all')
```

```python
sns.distplot(data_4['Year'])
plt.show()
```

```python
data_4
```

```python
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned
```

```python
plt.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
plt.xlabel('EngineV')
plt.ylabel('Price')
plt.show()

```

```python

plt.scatter(data_cleaned['Year'],data_cleaned['Price'])
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1,3,sharey=True,figsize=(12,4))
ax1.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax1.set_title('Price vs Mileage')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price vs EngineV')
ax3.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax3.set_title('Price vs Year')
plt.show()
```
