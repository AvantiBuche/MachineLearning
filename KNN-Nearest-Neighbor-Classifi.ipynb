```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
x=dataset.iloc[:,[2,3]].values
y=dataset.iloc[:,4].values
```

```python
plt.scatter(dataset["Age"],dataset["EstimatedSalary"],c=dataset["Purchased"])
plt.show()
```

```python
#train and split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.shape
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=8)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#KNN Classifier on iris data
from sklearn.datasets import load_iris
iris = load_iris()
data_iris = pd.DataFrame(data=iris.data,columns=iris.feature_names)
data_iris['label'] = iris.target
data_iris
```

```python
m = data_iris.iloc[:,0:4].values
n = data_iris.iloc[:,4].values
from sklearn.model_selection import train_test_split
m_train,m_test,n_train,n_test = train_test_split(m,n,test_size=0.2,random_state=0)

from sklearn.neighbors import KNeighborsClassifier
model_ir=KNeighborsClassifier(n_neighbors=8)
model_ir.fit(m_train,n_train)
n_pred=model_ir.predict(m_test)
n_pred
```

```python
n_test
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(n_test,n_pred))
```

```python
print(classification_report(n_test,n_pred))
```
