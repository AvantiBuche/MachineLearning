```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
X = iris.data
X
```

```python
# features Extraction
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_reduce = pca.fit_transform(X)
X_reduce
```

```python
y = iris.target
y
```

```python
plt.scatter(X_reduce[:,0],X_reduce[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
pca = PCA(n_components=3)
```

```python
X_reduce1 = pca.fit_transform(X)
```

```python
X_reduce1
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_reduce1[:,0],X_reduce1[:,1],X_reduce1[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
#Feature Selection

# Variance
# going to remove low variance features
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0.5)
X_vt = sel.fit_transform(X)
X_vt
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_vt[:,0],X_vt[:,1],X_vt[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
sel = VarianceThreshold(threshold=0.6)
X_vt = sel.fit_transform(X)
X_vt
```

```python
plt.scatter(X_vt[:,0],X_vt[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
# Select from model

from sklearn.svm import SVC
from sklearn.feature_selection import SelectFromModel
clf = SVC(kernel='linear')
clf.fit(X,y)
```

```python
model = SelectFromModel(clf,prefit=True)
X_sfm = model.transform(X)
X_sfm
```

```python
plt.scatter(X_sfm[:,0],X_sfm[:,1],c=y,cmap='rainbow')
plt.show()
```
