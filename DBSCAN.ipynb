```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

#create dmmy data
from sklearn.datasets import make_blobs
centers = [[0.5,2],[-1,-1],[1.5,-1]]

#skip Y here coz we don't use target in clustering
X,_ = make_blobs(n_samples=400,centers=centers,cluster_std=0.5,random_state=0)
X
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1])
plt.show()

from sklearn.cluster import DBSCAN
#eps= radius, min_samples=min point in a circle
db = DBSCAN(eps=0.4,min_samples=17)
db.fit(X)
```

```python
y_pred = db.fit_predict(X)
y_pred
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='rainbow')
plt.show()
#purple color are outliers/noises
```
